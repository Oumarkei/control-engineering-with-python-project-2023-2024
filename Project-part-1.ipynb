### Introduction

We consider a simple model of a self-balancing scooter evolving in the plane. It is mainly made of

  - The frame that contains the two articulated pads; from above, each pad essentially look like a triangle (one vertice in the center of the frame, one vertice at each extremity of the wheel).

  - Two fixed wheels (their orientation with respect to the frame does not change). 
    They are parallel and located at opposite ends of the frame. 
    
Let $(0, e_x, e_y)$ be a direct orthonormal frame of the plane. We denote 

  - $(x, y)$ the coordinates of the wheel axis center in the orthonormal and 
  - $\theta$ the angle of the wheels direction with respect to vector $e_x$ ; with $\theta = 0$ when the front of the scooter faces the right ($e_x$) and $\theta = \pi/2$ when it faces upwards ($e_y$).

  - $\phi_l$ and $\phi_r$ the rotation angles of the left and right wheel respectively ; by convention, when these angles increase, the scooter moves forward.

üß© Implement a function `draw_scooter` with arguments `x`, `y` and `theta` that draws a simple geometric model of the wheelchair (with a red frame and black fixed wheels) in the plane (top view) for arbitrary values of $(x, y)$ and $\theta$.
```python
# Let's define the rotation  function
def rotate(points, theta, origin):
    """Rotate points by theta radians around the origin."""
    R = np.array([[np.cos(theta+3*pi/2), -np.sin(theta+3*pi/2)], 
                  [np.sin(theta+3*pi/2),  np.cos(theta+3*pi/2)]])
    return (points - origin) @ R.T + origin
```

```python
eps = l/20
def draw_scooter(x, y, theta) :
    
    # Centre de rotation
    origin = np.array([x, y])
    
    # The coordinates of rectangle vertex, 
    point1 = array([[x,y], [x+l/2, y-r], [x+l/2, y+r], [x,y]])
    wheel1 = array([[x + l / 2 - eps, y - r], [x + l / 2 + eps, y - r], 
                       [x + l / 2 + eps, y + r], [x + l / 2 - eps, y + r], 
                       [x + l / 2 - eps, y - r]])
    point2 = array([[x,y], [x-l/2, y+r], [x-l/2, y-r], [x,y]])
    
    wheel2  = array([[x - l / 2 - eps, y - r], [x - l / 2 + eps, y - r], 
                       [x - l / 2 + eps, y + r], [x - l / 2 - eps, y + r], 
                       [x - l / 2 - eps, y - r]])
    
    # Appliquer la rotation aux points
    point1 = rotate(point1, theta, origin)
    point2 = rotate(point2, theta, origin)
    wheel1 = rotate(wheel1, theta, origin)
    wheel2 = rotate(wheel2, theta, origin)
    
    # Position du point scatter (vert) au-dessus du centre de rotation
    scatter_point = np.array([x, y + 0.5])
    scatter_point = rotate(scatter_point, theta, origin)
    
    #Let's plot the first triangle
    fill(point1[:, 0], point1[:, 1], 'red')
    
    # Let's plot the second triangle
    fill(point2[:, 0], point2[:, 1], 'red')
    
    # Let's draw the wheeels
    fill(wheel1[:, 0], wheel1[:, 1], 'black')
    fill(wheel2[:, 0], wheel2[:, 1], 'black')
    
    # Ajouter le point scatter au-dessus du centre de rotation
    scatter(scatter_point[0], scatter_point[1], c='g')
```

```python
draw_scooter(0, 0, 3*pi/2)
```
